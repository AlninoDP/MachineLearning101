```python
import matplotlib.pyplot as plt

rasa = ('Coklat', 'Strawberry', 'Blueberry', 'Orange', 'Mango' , 'Lemon')
vote = (22, 8, 15, 25, 16, 4)
explode = (0, 0, 0, 0.1, 0.2, 0)

plt.pie(vote, labels= rasa, autopct= '%1.1f%%', explode=explode,shadow=True)
```
