3. Taxonomi dalam machine Learning:

Berdasarkan Jenis Tugas (Task-based):
- Supervised Learning: Model dilatih dengan menggunakan data yang berisi pasangan input dan output.
- Unsupervised Learning: Model belajar dari data tanpa label, mencoba mengidentifikasi pola atau struktur di dalamnya.
- Reinforcement Learning: Model belajar melalui interaksi dengan lingkungan, menerima umpan balik dalam bentuk reward atau punishment.

Berdasarkan Tujuan (Goal-based):
- Classification (Klasifikasi): Memprediksi kelas atau label dari suatu data.
- Regression: Memprediksi nilai kontinu atau rentang nilai.
- Clustering: Mengelompokkan data ke dalam kelompok-kelompok yang serupa. 
- Dimensionality Reduction: Mengurangi jumlah fitur atau atribut dalam data.
- Anomaly Detection: Mengidentifikasi pola yang tidak biasa atau anomali dalam data.

Berdasarkan Representasi Model:
- Symbolic Learning: Mewakili pengetahuan dalam bentuk aturan atau simbol.
- Connectionist Learning: Mewakili pengetahuan melalui jaringan saraf tiruan.

Berdasarkan Metode Pembelajaran (Learning Paradigms):
- Batch Learning: Model dilatih pada seluruh dataset sekaligus.
- Online Learning: Model diperbarui secara berkelanjutan saat data baru tiba.
- Transfer Learning: Mentransfer pengetahuan dari satu tugas ke tugas lainnya.
- Meta-Learning: Model dilatih untuk belajar bagaimana belajar.

Berdasarkan Algoritma atau Teknik:
- Decision Trees: Model yang menggunakan struktur pohon keputusan.
- Support Vector Machines (SVM): Model yang membangun hyperplane untuk memisahkan kelas.
- Neural Networks: Model yang terinspirasi oleh struktur dan fungsi otak manusia.
- K-Nearest Neighbors (KNN): Model yang memprediksi berdasarkan mayoritas kelas tetangga terdekat.
- Clustering Algorithms: Seperti K-Means, Hierarchical Clustering, untuk mengelompokkan data.

Berdasarkan Sumber Data:
- Text Mining: Analisis dan ekstraksi informasi dari data teks.
- Image Processing: Analisis dan interpretasi data gambar.
- Speech Recognition: Pengenalan dan pemahaman ucapan manusia.
