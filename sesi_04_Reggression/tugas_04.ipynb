```python
import numpy as np
import pandas as pd

dataset = pd.read_csv('multi_linear_regression/Dataset.csv')

X = dataset.iloc[:, :-1] 
y = dataset.iloc[:, -1]
y
 
```

Encoding Categorical Data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
transformer = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')
X = np.array(transformer.fit_transform(X)) 
```

Splitting Dataset


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state=0) 
```

Training Model

```python
### Method fit() digunakan untuk training model


from sklearn.linear_model import LinearRegression
modelRegression = LinearRegression()
modelRegression.fit(X_train, y_train)
### X_train adalah data yang digunakan untuk melatih model atau x adalah data independen
### y_train adalah nilai yang ingin diprediksi berdasarkan X_train. atau y adalah data dependen
```

Evaluasi Model Yang Dibuat


```python
### Method pred() digunakan untuk prediksi/estimasi dengan memanfaatkan model yang di train 
### Fungsi predict() digunakan untuk memperoleh prediksi berdasarkan data uji yang dimasukkan ke dalam model.

y_pred = modelRegression.predict(X_test) ### variable y_pred berisi hasil prediksi nilai y berdasarkan X_test

y_pred_reshaped = y_pred.reshape(len(y_pred), 1)  # Mengubah y_pred menjadi bentuk kolom
y_test_reshaped = y_test.values.reshape(len(y_test), 1)  # Mengubah y_test menjadi bentuk kolom

np.set_printoptions(precision=2)
print(np.concatenate((y_pred_reshaped, y_test_reshaped),axis=1)) 
```

Gunakan Trained Modelnya Untuk Memprediksi Data Profit (data y)

```python
### 1,0,0 adalah california
### R&D spent = 160000
### Administrative spent = 130000
### Marketing Spent = 300000
newData = [[1,0,0, 160000, 130000, 300000]] 

profitPrediction = modelRegression.predict(newData)

print(profitPrediction)
```

Mengetahui seberapa kuat hubungan antara variabel-variabel 

```python
### Koefisien adalah kemiringan garis regresi yang ditemukan oleh model untuk setiap fitur dalam data.
### Intercept adalah titik di mana garis regresi memotong sumbu y saat x = 0.

print(modelRegression.coef_) 
print(modelRegression.intercept_)

### print(regressor.coef_) akan menunjukkan seberapa besar pengaruh setiap variabel pada prediksi
### print(regressor.intercept_) akan menunjukkan titik di mana garis lurus memotong sumbu vertikal.
```
